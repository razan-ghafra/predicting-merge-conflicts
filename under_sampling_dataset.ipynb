```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
from imblearn.under_sampling import RandomUnderSampler
```

```python
df = pd.read_csv("merged_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# feature extraction
X = df.drop(columns=["Unnamed: 0","pr_id","ms_id", "has_conflict"])  # Remove non-feature columns
y = df["has_conflict"]  # Target column
```

```python
# random under sampling
rus = RandomUnderSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)
```

```python
df_rus = pd.DataFrame(X_resampled, columns=df.drop(["Unnamed: 0","pr_id","ms_id",'has_conflict'], axis=1).columns)
df_rus['has_conflict'] = y_resampled
```

```python
df_rus.info()
```

```python
# Check the new class distribution
print(f"Resampled class distribution: {df_rus['has_conflict'].value_counts()}")
```

```python
# save the dataframe
df_rus.to_csv('RUN_dataset.csv', index=False)
```
